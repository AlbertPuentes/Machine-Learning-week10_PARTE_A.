```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv('usuarios_1000.csv')

# 1. Demostrar el uso de head(), info(), describe()
print("=== 1. HEAD ===")
print(df.head())
print("\n=== 1. INFO (Información del DataFrame) ===")
df.info()
print("\n=== 1. DESCRIBE ===")
print(df.describe())

# 2. Seleccionar la columna "edad"
print("\n=== 2. COLUMNA EDAD ===")
edad_col = df['edad']
print(edad_col)

# 3. Filtrar ingresos mayores a 3000
print("\n=== 3. Clientes con ingreso > 3000:")
filtro = df[df["ingreso"] > 3000]
print(filtro)

# 4. Análisis por categorías usando groupby (Promedio por ciudad)
print("\n=== 4. ANÁLISIS POR CIUDAD USANDO GROUPBY ===")
analisis_ciudad = df.groupby('ciudad').mean(numeric_only=True)
print(analisis_ciudad)

# 5. Ordenamiento usando sort_values()
print("\n=== 5. ORDENAMIENTO POR INGRESO ===")
df_ordenado = df.sort_values(by='ingreso', ascending=False)
print(df_ordenado)

# 6. Almacenar los resultados usando to_csv()
df_ordenado.to_csv('usuarios_otro.csv', index=False)
print("\n=== 6. ARCHIVO EXPORTADO ===")
print("El archivo 'usuarios_otro.csv' ha sido creado exitosamente.")

# 7. Clientes con edad mayor a 30 años
print("\n=== 7. CLIENTES MAYORES A 30 AÑOS ===")
clientes_ma_30 = df[df['edad'] > 30]
print(clientes_ma_30)

# 8. Mostrar el cliente con mayor gasto
print("\n=== 8. CLIENTE CON MAYOR GASTO ===")
mayor_gasto = df[df['gasto_mensual'] == df['gasto_mensual'].max()]
print(mayor_gasto)

# 9. Calcular el gasto promedio
gasto_promedio = df['gasto_mensual'].mean()
print(f"\n=== 9. GASTO PROMEDIO ===\nEl gasto promedio es: {gasto_promedio:.2f}")

# 10. Calcular el ingreso promedio
ingreso_promedio = df['ingreso'].mean()
print(f"\n=== 10. INGRESO PROMEDIO ===\nEl ingreso promedio es: {ingreso_promedio:.2f}")

# 11. Analizar dónde gastan más los clientes por ciudad
print("\n=== 11. GASTO PROMEDIO POR CIUDAD ===")
gasto_por_ciudad = df.groupby('ciudad')['gasto_mensual'].mean().sort_values(ascending=False)
print(gasto_por_ciudad)
```
